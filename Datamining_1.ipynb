```python
#important to normalise or standardise the data at this point since there are 
#differences in the ranges of the data and how they are measured

#define min max scaler function with feature as a parameter
def min_max_scaler(feature):
  scaled_list = []
  for attribute in feature:
    scaled_list.append((attribute - min(feature))/(max(feature) - min(feature)))
  scaled_list
  return(scaled_list)
```

# Part 2 - k-NN implementation


Found example code for calculating Euclidean distance in NumPy from here:

```python
#we can also find euclidean distance using np.linalg.norm(series1, series2)
#either way is fine

#Euclidean Distance
def Euclidean_distance(a, b):
  distance = np.sqrt(np.sum(np.square(a - b)))
  return(distance)

Euclidean_distance(test.iloc[1,:], train.iloc[1,:])
```

Found a tutorial online to follow to get the basic structure of a k-NN algorithm:

https://www.youtube.com/watch?v=iJfcRV4PPnY

Needed a mode function for this to work but we are not allowed to use scipy so I
found a mode function to adapt from the internet: 

https://stackoverflow.com/questions/12399107/alternative-to-scipy-mode-function-in-numpy

Edited so that it does not return oldcounts because they are not necessary here.

For our first test we will use a hyperparameter of 3.

```python
def mode(a, axis=0):
    scores = np.unique(np.ravel(a))
    test_shape = list(a.shape)
    test_shape[axis] = 1
    old_most_freq = np.zeros(test_shape)
    old_counts = np.zeros(test_shape)

    for score in scores:
        template = (a == score)
        counts = np.expand_dims(np.sum(template, axis),axis)
        most_frequent = np.where(counts > old_counts, score, old_most_freq)
        old_counts = np.maximum(counts, old_counts)
        old_most_freq = most_frequent

    return most_frequent

#k-NN algorithm:
def k_NN(train_x, test_x, train_y, test_y, k):
  predictions = []
  for test_pt in test_x.to_numpy():
    distances = []
    for i in range(len(train_x)):
      distances.append(Euclidean_distance(np.array(train_x.iloc[i]), test_pt))

    distance_data = pd.DataFrame(data = distances, columns = ["distance"],
                                 index = train_y.index)
    
    k_neighbours_list = distance_data.sort_values(by = ["distance"],
                                                  axis = 0)[:k]

    labels = train_y.loc[k_neighbours_list.index]

    voting = mode(labels)[0]

    predictions.append(voting)
  return(predictions)

predict_list = k_NN(train_data, test_data, train_labels, test_labels, 3)

```

```python
#we have to remember that because we took the slice of test_labels off the end
#of the original dataframe, the indexes start at 710.
def accuracy_score(predict_list, test_labels):
  count = 0
  for score_pos in range(len(test_labels)):
    if test_labels[score_pos + 710] == predict_list[score_pos]:
      count += 1
  accuracy_score = count/(len(test_labels))
  return(accuracy_score)
```

```python
accuracy_score(predict_list, test_labels)
```

```python
def confusion_matrix(predict_list, test_labels):
  
  true_positives = 0
  false_positives = 0
  true_negatives = 0
  false_negatives = 0

  confusion_matrix = pd.DataFrame({"": ["Predict_Survived", "Predict_Died"],
                                   "Test_Survived": [0,0],
                                   "Test_Died": [0,0]})

  for score_pos in range(len(test_labels)):
    if test_labels[score_pos + 710] == 1:
      if test_labels[score_pos + 710] == predict_list[score_pos]:
        true_positives += 1
      else:
        false_negatives += 1
    else:
      if test_labels[score_pos + 710] == predict_list[score_pos]:
        true_negatives += 1
      else:
        false_positives += 1
    
  confusion_matrix.iloc[0,1] = [true_positives]
  confusion_matrix.iloc[0,2] = [false_positives]
  confusion_matrix.iloc[1,1] = [false_negatives]
  confusion_matrix.iloc[1,2] = [true_negatives]

  return(confusion_matrix)
```

```python
confusion_matrix(predict_list, test_labels)
```

The confusion matrix reflects that the accuracy score of the k-NN algorithm we have created is high. We can see that the number of true positives (target variable survived both in the test labels and predictions) is 53 and the number of true negatives 99. 53 + 99 / 177 = ~85% which is the accuracy score that we saw earlier.

# Part 4 - Weighted k-NN 




I was using a mode approach for normal k-NN, but here it makes more sense to let the higher weight decide which outcome should be predicted. Therefore I modified our code to include positive and negative weight totals using a dataframe that contained the calculated weights and distances.

```python
print(test_labels.index)
```

```python
#weighted k-NN algorithm:
def weighted_k_NN(train_x, test_x, train_y, test_y, k):
  predictions = []
  for test_pt in test_x.to_numpy():
    
    distances = []
    weights = []
    pos_weights_list = []
    neg_weights_list = []
    pos_weight_total = 0
    neg_weight_total = 0

    for i in range(len(train_x)):
      distances.append(Euclidean_distance(np.array(train_x.iloc[i]), test_pt))

    distance_data = pd.DataFrame(data = distances, columns = ["distance"],
                                 index = train_y.index)

    k_neighbours_df = distance_data.sort_values(by = ["distance"])[:k]

    for i in k_neighbours_df["distance"]:
      #code for if i == 0 explained in comments
      if i == 0:
        weights.append(1 / np.square(0.0000001))
      else:
        weights.append(1 / np.square(i))
    
    k_neighbours_df["weights"] = weights

    for i in k_neighbours_df.index:
      if train_y.loc[i] == 1:
        pos_weights_list.append(k_neighbours_df.loc[i, "weights"])
        pos_weight_total = np.sum(pos_weights_list)
      else:
        neg_weights_list.append(k_neighbours_df.loc[i, "weights"])
        neg_weight_total = np.sum(neg_weights_list)
    
    #code for a tie explained in comments
    if pos_weight_total == neg_weight_total:
      if len(pos_weights_list) > len(neg_weights_list):
        predictions.append(0)
      else:
        predictions.append(1)   
    elif pos_weight_total > neg_weight_total:
      predictions.append(1)
    else:
      predictions.append(0)

  return(predictions)

weighted_predict_list = weighted_k_NN(train_data,
                                      test_data,
                                      train_labels,
                                      test_labels, 3)
```

#Comments on results:
It is interesting to note that excluding the case where the hyperparameter is 1 (which of course gives lower accuracy due to not having enough deciding factors), here the accuracy score and confusion matrices stayed nearly exactly the same throughout the testing of multiple hyperparameters. This makes sense as using distance weights to decide the predictions means the size of the k being used is less important. This is because as the hyperparameter grows larger it must include data points further away from the target. However, these points are given less importance because their distance away from the target is factored in.

#In the case of i == 0 
Since i is basically the same point as the target variable it clearly holds weight. To solve the division by 0 problem I have set the value to a miniscule  value of 0.0000001 to give a high weight.

#In the case of a tie:
The list with less weights is given precedence because there may be less weights but they are of a higher value each. For example, there could be a list with 1 weight that has a value of 10 and the other list has 10 weights that each have values of 1. The closest neighbour here is definitely the neighbour with a value of 10 and in my opinion it is more likely to be an accurate predictor of the outcome of the target variable.

#k-NN or Weighted k-NN?

The k-NN algorithm performed better than the weighted k-NN algorithm for this dataset with all values of k that we tested with regards to accuracy score and results of the confusion matrices. Therefore it is perhaps better to use the k-NN algorithm for this specific dataset if we are using a k value of around 5. However, if we are going to use a larger k value then I believe we should use the weighted k-NN algorithm as it has proven to be more constant in accuracy results when it comes to a changing hyperparameter. It is also worth noting that the difference between an 83% accuracy score and a range of [84%,87%] accuracy score is not an extreme difference. Both of the algorithms performed well and produced accurate results.
